```
trainX = trainX / 255.0 # Normalize # Standardize
# hiện tại đang chuẩn hóa theo kiểu chia tất cả các giá trị trong ma trận cho giá trị max, ngoài ra có thể dùng cách khác
```

```
trainX[0]
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
```

```
trainX.shape
```

```
import numpy
trainX.reshape((trainX.shape[0], -1)).shape
```

```
# Target: (60000, 28, 28) -> (60000,)
```

```
# 1. Tạo mạng Sequential
model = Sequential()
# 2. Thêm lớp Flatten() # 28 * 28 thành vector có chiều (784,)
model.add(Flatten(input_shape=(28, 28))) # (28, 28) -> (784, 1)
# 3. Thu về miền của nhãn
model.add(Dense(10, activation='softmax'))
```

```
model.summary()
```

Bản chất của các code trên là lập trình sao cho mỗi data đi qua các lớp trên r trả ra kết quả. Như trên thì ví dụ 1 ảnh 28x28 sẽ đi qua lớp flatten để duỗi về (784,1), sau đó đi qua lớp dense sotfmax để train và sau đó trả ra kết quả

```
trainY[0]
```

```
model.compile(optimizer='sgd', loss='sparse_categorical_crossentropy', metrics=['acc'])
```

sparse_categorical_crossentropy mục địch là chuyển về vector onehot để từ đó nhân hadamat để hình thành softmax

chia tập dữ liệu thành 32 batch nhỏ để tiện cho quá trình train -> mỗi batch sẽ phải chạy 1875 step

```
model.fit(trainX, trainY, epochs=10, batch_size=32, validation_data=(testX, testY))
```

```
model.save('model.h5')
```
